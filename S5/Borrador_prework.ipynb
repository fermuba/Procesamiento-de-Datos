### Diccionario

- Suburb: Nombre del suburbio donde se encuentra la propiedad.
- Address: Dirección específica de la propiedad.
- Rooms: Número de habitaciones en la propiedad.
- Type: Tipo de propiedad. Generalmente, esto puede incluir categorías como:
    + 'h' para casa (house)
    + 'u' para unidad (unit)
    + 't' para townhouse (casa adosada)
- Price: Precio de venta de la propiedad.
- Method: Método de venta. Algunos ejemplos comunes incluyen:
    + 'S' para venta estándar
    + 'SP' para venta con precio
    + 'PI' para venta antes de la subasta
    + 'VB' para ventas de oferta y negociación
    + 'SA' para ventas después de la subasta
- SellerG: Nombre del agente o agencia inmobiliaria que vende la propiedad.
- Date: Fecha de la venta de la propiedad.
- Distance: Distancia de la propiedad al centro de la ciudad (en kilómetros).
- Postcode: Código postal donde se encuentra la propiedad.
- Bedroom2: Número de dormitorios secundarios en la propiedad (puede ser redundante con 'Rooms' en algunos casos).
- Bathroom: Número de baños en la propiedad.
- Car: Capacidad de estacionamiento de la propiedad (número de coches que se pueden estacionar).
- Landsize: Tamaño del terreno en metros cuadrados.
- BuildingArea: Área construida en metros cuadrados.
- YearBuilt: Año de construcción de la propiedad.
- CouncilArea: Área del consejo municipal en la que se encuentra la propiedad.
- Lattitude: Latitud geográfica de la propiedad.
- Longtitude: Longitud geográfica de la propiedad.
- Regionname: Nombre de la región o zona en la que se encuentra el suburbio.
- Propertycount: Número total de propiedades en el suburbio.

### Aplicación en el dataset del inicio

```python
# Vamos a leer un archivo .csv para poder trabajar en el
df = pd.read_csv('../Datasets/melbourne_housing-raw.csv', sep= ',')

# Vemos las primeras 5 filas
df.head()     # por defecto lee las 5, podemos cambiar el valor
```

```python
# Contamos los NaN de cada columna
df.isna().sum()
```

```python
# Eliminamos las columnas con un numero elevado de NaN
df_2 = df.drop(columns= ['BuildingArea', 'YearBuilt'])

df_2.isna().sum()
```

```python
# Rellenamos 'Regionname' con Unknown
df_2['Regionname'] = df['Regionname'].fillna('Unknown')

df_2.isna().sum()
```

```python
# Ahora eliminamos las filas donde haya elementos sean NaN
df_dropped = df_2.dropna(axis = 0, how= 'any')

df_dropped.isna().sum()
```

```python
# El tamaño nuevo es
df_dropped.shape
```

```python
# Al eliminar tantas filas, nuestro indice no es continuo para reasinar uno continuo usamos
df_dropped.reset_index(inplace= True)

# Vemos los ultimos elementos
df_dropped.tail()
```

```python
# Para renombrar columnas  vamos a usar rename, necesitamos un diccionario
# con las palabras que queremos reemplazar y las nuevas
column_name_mapping = {
    'Suburb': 'suburb',
    'Address': 'address',
    'Rooms': 'rooms',
    'Type': 'type',
    'Price': 'price',
    'Method': 'method',
    'SellerG': 'seller_g',
    'Date': 'date',
    'Distance': 'distance',
    'Postcode': 'post_code',
    'Bedroom2': 'bedrooms',
    'Bathroom': 'bathroom',
    'Car': 'car',
    'Landsize': 'land_size',
    'CouncilArea': 'council_area',
    'Lattitude': 'latitude',
    'Longtitude': 'longitude',
    'Regionname': 'region_name',
    'Propertycount': 'property_count'
}
```

```python
# Aplicamos el reemplazo con el diccionario anterior
df_renamed = df_dropped.rename(columns= column_name_mapping)

df_renamed
```
